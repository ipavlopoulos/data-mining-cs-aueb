## Pandas Axes

* In pandas, `DataFrames` exist on two axes.

* The row axis, going vertically, is axis 0.

* The column axis, going horizontally, is axis 1.

* So, when we do an operation on an entire row, we are going through axis 1.

* Similarly, when we do an operation on an entire column, we are going through axis 0.

# Using the `[]` operator

* The most obvious way to select stuff with pandas is with the `[]` operator.

* This provides familiarity with Python.

* However, it is usually not the optimal way.

* We do better by using more specialised methods.

## Selecting a Single Column

* We can select a single column by entering its name in `[]`.

* This yields a `Series` object.

* `df['A']` is equivalent to `df.A`.

* Setting the scene:

````python
import pandas as pd
import numpy as np

dates = pd.date_range('20160101', periods=6)

dates
````

````python
df1 = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list('ABCD'))

df1
````

````python
df1['A']
````

## Selecting Multiple Columns

* If we give a list of column names, we will get a dataframe with these columns.

````python
df1[['A', 'C']]
````

## Selecting Rows

* If we pass a slice to `[]`, we will get a slice containing these rows.

* The right end is not included, as in standard Python.

````python
df1[1:3]
````

## Columns with Numeric Names

* What will happen if our columns contain numeric names?

````python
df2 = pd.DataFrame(np.random.randn(6,4), index=dates, columns=range(4))

df2
````

* When we pass a single number to `[]` we will select that column and will get the corresponding `Series`.

````python
df2[2]
````

* But if we pass a slice to `[]`, we will slice the rows.

````python
df2[1:3]
````

* If we pass a list to `[]`, we will get the corresponding columns.

````python
df2[[1, 2]]
````

* The situation remains the same if both our columns and our index are numeric.

````python
df3 = pd.DataFrame(np.random.randn(6,4), columns=range(4))

df3
````

* Providing a slice to `[]` slices the *rows*, so that the right (i.e., the lower end; we are slicing vertically) end is *not* included.

````python
df3[1:3]
````

* Note that all slice variations are supported, for example:

````python
df3[1:6:2]
````

* Providing a list to `[]` selects the *columns*.

````python
df3[[1, 3]]
````

* We can also use a boolean array to select the rows meeting the condition:

````python
df1[df1['B'] > 0]
````

#### Takeaways regarding `[]`

* With a DataFrame, slicing inside of `[]` *slices the rows*.

* With a DataFrame, passing a list (or an item) inside of `[]` *selects the columns*.

## Selection by *label* with `.loc`

* A *label* in pandas refers to the *index*.

* If the index is numeric, it may look like we are selecting based on the corresponding positions, *but it is not so*.

* There must be at least *one* row selected.

* In contrast to standard Python, both the *start* and *the end* are included in slices.


* Selection by label is performed with the `.loc` method.
* It accepts a:
    * **single label**, such as the number 5 or the label `'a'`. 
        * The number (e.g., 5) is interpreted as a label of the index (i.e., `'5'`). 
        * It is *not* a numeric position on the index. 
        * If the index is numeric, it just happens to look so.
    * **list or array** of labels, such as `['a', 'b', 'c']`.
    * **slice object** with labels `'a':'f'`.
        * Both ends are included!
        * Slice objects can be of the form `x:y:z`.
    * **boolean array**.
    * **callable** i.e., a function.

* Setting the scene:

````python
import numpy as np
df1 = pd.DataFrame(np.random.randn(6,4),
                   index=list('abcdef'),
                   columns=list('ABCD'))

df1
````

* Accessing the *row* via a single label:

````python
df1.loc['a']
````

* Note that a label is always a label, not a position, even if it happens to be numeric.

* To see that, let us create a frame with numeric indices:

````python
df2 = pd.DataFrame(np.random.randn(6, 4),
                  index=range(5, -1, -1),
                  columns=list('ABCD'))
df2
````

````python
df2.loc[1]
````

* Back to `df1` for a list of labels:

````python
df1.loc[['b', 'e']]
````

## Inclusion
* A slice where *both ends are included*:

````python
df1.loc['b':'e']
````

* We can select both rows and columns. The columns are separated from the rows by `,`.

````python
df1.loc['b':'e', # <== rows
        'B':'D'] # <== columns
````

## All rows but few columns
* If we want a projection on columns, containing all the rows, we use `:` for the rows:

````python
df1.loc[:, 'B':'D']
````

* Slicing treats the rows and the columns in their existing order in the data source.

* It **does not re-order them** in any way.

* Let's create a `DataFrame` with some sorted data to see what this means.

````python
import pandas as pd
updown = pd.DataFrame({
    'A': range(1, 6),
    'X': range(11, 16),
    'B': range(6, 1, -1)},
    index=list('abxde'))
updown
````

````python
updown.sort_values(by='B', inplace=True)
updown
````

* Now, if we take the slice from `b` to `d`, we'll just get an empty `DataFrame`:

````python
updown.loc['b':'d']
````

* The slice from `d` to `b` will return the rows that appear in that interval.

* These *include* the `x` row.

````python
updown.loc['d':'b']
````

* Similarly, a slice from the `A` column to the `B` column will include the `X` column:

````python
updown.loc['d':'b', 'A':'B']
````

## Selection by position: `.iloc`

* Selection by position means pure integer-based addressing.

* Positions start from 0 and their end is not included, as in standard Python (from 0 to length-1 of the axis).

* Selection by position in pandas is with the `.iloc` method.


* It accepts the following:
    * An integer, e.g., 5.
    * A list or array of integers, such as `[4, 3, 0]`.
    * A slice object with ints, like `1:7`.
    * Slice objects can be of the form `x:y:z`.
    * A boolean array.
    * A callable.

* Setting the scene.
* To make it obvious that a position is not the same as the label, we'll have integer labels, which in general will not coincide with the positions.

````python
df1 = pd.DataFrame(np.random.randn(6,4),
                   index=list(range(0,12,2)),
                   columns=list(range(0,8,2)))
df1
````

* Accessing via an *integer* selects the **row** in that position:

````python
df1.iloc[2]
````

* Selecting with an *integer list* selects the **rows** in the respective positions:

````python
df1.iloc[[1,3]]
````

* Similarly with columns:

````python
df1.iloc[:, [1, 3]]
````

* And both rows and columns:

````python
df1.iloc[[1,3], [1,3]]
````

* Selecting via integer slicing selects the rows in the slice:

````python
df1.iloc[2:4]
````

* If we want the selection only on the columns, we just specify that we want all the rows:

````python
df1.iloc[:, 1:3]
````

## Boolean Indexing

* Boolean indexing is used for filtering our data based on a particular condition.

* Conditions may be combined with boolean operators. In that case, *each condition must be put in parentheses*.

* The boolean operators are `~` (not), `&` (and), `|` (or).

* Setting the scene:

````python
df1 = pd.DataFrame(np.random.randn(6,5), index=dates, columns=list('ABCDE'))
df1
````

* Boolean AND

````python
df1[(df1['A'] > 0) & (df1['A'] < 1)]
````

* Boolean OR:

````python
df1[(df1['A'] > 2) | (df1['A'] < 0)]
````

* Boolean NOT:

````python
df1[~((df1['A'] > 2) | (df1['A'] < 0))]
````

## Selection by Callable

* We can also use a callable for indexing.

* One that receives one argument and returns a value that can be used for indexing.

* E.g., `True` or `False` for rows and a column selector for columns.

* Setting the scene:

````python
df1 = pd.DataFrame(np.random.randn(6, 4),
                   index=list('abcdef'),
                   columns=list('ABCD'))
df1
````

* A callable on rows:

````python
df1.loc[lambda df: df.A > 0, :]
````

* A callable on rows and columns:

````python
df1.loc[lambda df: df.A > 0, lambda df: ['C', 'D']]
````

* A callable on rows, by position:

````python
df1.iloc[lambda df: [0, 1], :]
````

* A callable on rows and columns, by position:

````python
df1.iloc[lambda df: [0, 1], lambda df: [2, 3]]
````

## Multi-dimensional Indexing

* Pandas also supports multi-dimensional indexing, where an index is essentially a hierarchical structure.

* That means that we index first by the first dimension, then by the second, etc.

* As an example, we are going to use the [Airline On-Time Performance Data](https://www.transtats.bts.gov/Tables.asp?DB_ID=120&DB_Name=Airline%20On-Time%20Performance%20Data&DB_Short_Name=On-Time#), maintained by the [Bureau of Transportation Statistics (BTS)](https://www.bts.gov/) of the U.S. Department of Transportation.

* In particular, we'll use the Reporting Carrier On-Time Performance (1987-Present) dataset.

* It contains: "Monthly data reported by US certified air carriers that account for at least one percent of domestic scheduled passenger revenues--includes scheduled and actual arrival and departure times for flights".

* We'll use all the data for all the months of a single year. You can create this subset in [this online notebook](https://colab.research.google.com/drive/1GD_NAaLgIRQvAIDshATxJO_94MDpmYrV?usp=sharing)

* After processing and saving the file (e.g., `flights2020.csv.zip`), we can read it as follows:

````python
flights = pd.read_csv('flights2020.csv.zip', index_col=0)
print(flights.shape)
flights.sample(5)
````

## Sample
* Note that we used the `sample` method to get a subset of the `DataFrame`.

* The `sample` method can sample with or without replacement (parameter `replace`, default is `False`).

* We can also pass it a column of weights, indicating the probability that each row will get selected.

## Build a hierarchical index
* Now, let's create a hierarchical index using the following fields:
    * `CARRIER`
    * `ORIGIN`
    * `DEST`
    * `TAIL_NUM`
    * `FL_DATE`

````python
selection = ['Reporting_Airline', 'Origin',  'Dest', 'Tail_Number', 'FlightDate']

hierarchical_flights = flights.set_index(selection)
hierarchical_flights = hierarchical_flights.sort_index(level=selection)
hierarchical_flights
````

## MultiIndex
* A hierarchical index is of type `MultiIndex`. 

* The levels of a `MultiIndex` are the names of the columns we used for indexing.

* The labels are the values in a level, that is, the values of the respective column.

* We can refer to levels by name or position. Level 0 is the outermost level.

* We'll select the columns DepTime and DepDelay where the carrier was American Airlines, Delta, or United Airways.

````python
hierarchical_flights.loc[['AA', 'DL', 'UA'], ['DepTime', 'DepDelay']]
````

* Now, let's put the hierarchical index into work.

* We'll select the rows whose carrier was American Airlines, Delta, or United Airlines and the origin was Chicago O'Hare (ORD) or Des Moines International Airport (DSM). 

* We'll wrap the two levels of the index in a tuple to do the job.

````python
hierarchical_flights.loc[(['AA', 'DL', 'UA'], ['ORD', 'DSM']), ['DepTime', 'DepDelay']]
````

## IndexSlice
* Suppose now that we want all flights between these two airports (origin DSM or ORD), and not just from these two carriers.

* We can do that by using an `IndexSlice`.

````python
hierarchical_flights.loc[pd.IndexSlice[:, ['ORD', 'DSM']], 
                         ['DepTime', 'DepDelay']]
````

## Using more information from the index
* We can get even more creative.

* Remember that the last level of the index was the flight date.

* Suppose we want to find the flights in the first half of January from Chicago O'Hare (ORD) to Des Moines International Airport (DSM). We'll do:

````python
hierarchical_flights.loc[pd.IndexSlice[:, 'ORD', 'DSM', :, '2020-01-01':'2020-01-15'],
                         ['DepTime', 'DepDelay', 'ArrTime', 'ArrDelay']]
````

# Pivot

* The pivot operation creates new columns out of the values of a single column.

* Let's start setting the scene:

````python
import pandas as pd
import numpy as np

dates_range = pd.date_range('2016/10/1', periods=3).repeat(4)
dates_range
````

````python
statuses = np.tile(list('ABCD'), 3)
statuses
````

````python
values = np.random.randn(12)
values
````

````python
df = pd.DataFrame({'date': dates_range, 'status': statuses, 'value': values})
df.head(3)
````

* Now, if we want to operate on the variables, for example on the `status`, a better representation would be to turn the statuses to distinct columns.

* We can do that with the `pivot` method. It takes the following arguments:
    * `index`: The column that will become the new index.
    * `columns`: The column whose values will become the new columns.
    * `values`: The column whose values will fill the new dataframe.

````python
df.pivot(index='date', columns='status', values='value')
````

## Pivot Tables

* We can also pivot a table, yielding a dataframe with many dimensions.

* This can be created with the `pivot_table` function. It takes the following arguments:

    * `data`: A DataFrame object
    * `values`: A column or a list of columns to aggregate.
    * `index`: A column, or array, which has the same length as data, or list of them. Will be used as the index of the pivot table. 
    * `columns`: A column, or array which has the same length as data, or list of them. The values will be used to create the columns of the pivot table.
    * `aggfunc`: The function to use for aggregation, defaulting to `numpy.mean`.


````python
import datetime

df = pd.DataFrame({'A': ['one', 'one', 'two', 'three'] * 6,
                   'B': ['A', 'B', 'C'] * 8,
                   'C': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 4,
                   'D': np.random.randn(24),
                   'E': np.random.randn(24),
                   'F': [datetime.datetime(2016, i, 1) for i in range(1, 13)] +
                        [datetime.datetime(2016, i, 15) for i in range(1, 13)]})
df.head(10)
````

````python
# Aggregating on D, indexing on A and B, using as cols the values of C 
pv = pd.pivot_table(df, values='D', index=['A', 'B'], columns=['C'])
pv
````

* Note that this is a *hierarchical index*, comprising both columns A and B.

* You can think of each value of the hierarchical index as a tuple.

````python
pv.index
````

* But of course it is not necessary to have a hierarchical index as a result of a `pivot_table`.

````python
pv = pd.pivot_table(df, values='D', index=['B'], 
                    columns=['A', 'C'], aggfunc=np.sum)
pv
````

````python
pv.index
````

* And you can add multiple aggregation functions (by default, the mean is computed).

````python
pv = pd.pivot_table(df, values='D', index=['B'], 
                    columns=['A', 'C'], aggfunc={"D":[min, max]})
pv
````

# Stacking and Unstacking

* A related operation to pivoting is *stacking*, and its inverse, *unstacking*.

* The stack operation takes `DataFrame` columns and adds them to the (hierarchical) index as another component.

* The unstack operation takes a component from the hierarchical index and turns it to a column.

### Create a List of Tuples

* We'll start by creating a multi-level index.

* The index will be constructed by two-elements tuples, using Python's `zip()` function.

* `zip` takes as arguments iterables and aggregates (zipping) the elements from each one.

* If `a` and `b` are lists, `zip(a, b)` will produce the tuples `(a[0], b[0]), (a[1], b[1]), ...` until `a` or `b` are exhausted.

````python
tuples = list(zip(['bar', 'bar', 'baz', 'baz','foo', 'foo', 'qux', 'qux'],
                  ['one', 'two', 'one', 'two','one', 'two', 'one', 'two']))

tuples
````

* We will create our `multi-index` from these tuples.

````python
multi_index = pd.MultiIndex.from_tuples(tuples, names=["first", "second"])
multi_index
````

* And now we can setup a hierarchical view of the dataframe using this multi-indexing.

````python
import pandas as pd
df = pd.DataFrame(np.random.randn(8, 2), index=multi_index, columns=['A', 'B'])
df
````

* By calling `stack()`, now, we can add a third level to our multi-index, which will correspond to the columns `A` and `B`.

````python
stacked = df.stack()
stacked
````

````python
stacked.index
````

* The inverse operation is `unstack()`.

````python
unstacked = stacked.unstack()
unstacked
````

````python
unstacked.index
````

* Note that `unstack()` by default unstacks the last level, but we can choose differently if we want.

* The following is the same as the `stacked.unstack()` that we did before:

````python
unstacked = stacked.unstack(2)
unstacked
````

````python
unstacked.index
````

* But here we unstack a different column:

````python
unstacked = stacked.unstack(1)
unstacked
````

````python
unstacked.index
````

* And here we unstack the first column:

````python
unstacked = stacked.unstack(0)
unstacked
````

````python
unstacked.index
````

* A more advanced use of `stack` is when the columns also have a `MultiIndex`.

````python
columns = pd.MultiIndex.from_tuples([
        ('A', 'cat', 'long'), ('B', 'cat', 'long'),
        ('A', 'dog', 'short'), ('B', 'dog', 'short')],
        names=['exp', 'animal', 'hair_length'])

df = pd.DataFrame(np.random.randn(4, 4), columns=columns)
df
````

* By stacking, we move `hair_length` into the index, creating a MultiIndex with two indices. 

````python
df.stack()
````

* We can also select the levels of the column `MultiIndex` to stack based on their level names:

````python
df.stack(level=['animal', 'hair_length'])
````

* Or based on their level numbers:

````python
df.stack(level=[1, 2])
````
